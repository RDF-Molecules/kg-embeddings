```python
import pandas as pd
import numpy as np
from io import StringIO
%matplotlib inline
from rdflib import Graph,URIRef
from gastrodon import LocalEndpoint,one,QName
pd.set_option("display.width",100)
pd.set_option("display.max_colwidth",80)
import statistics
```

## Load data 

```python
input_path = "/media/rivas/Data1/Data-mining/Kcap/datasets/"

f_dbp_onto = input_path + "DBpedia ontology 2016/" + "dbpedia_2016-10.nt"
f_dbp_people = input_path + "DBpedia_people/" + "goldStandard_dbp_cleaned.nt"
f_drug = input_path + "Drugs/" + "goldStandard_drugs_copy.nt"

dbpedia_ontology = Graph()
dbpedia_ontology.parse(f_dbp_onto, format="nt")
dbpedia_People = Graph()
dbpedia_People.parse(f_dbp_people, format="nt")
drugbank = Graph()
drugbank.parse(f_drug, format="nt")
#drugbank.parse(f_drugbank, format="turtle")

print(len(dbpedia_ontology))
print(len(dbpedia_People))
print(len(drugbank))
```

## Merging graphs (dbpedia_ontology + dbpedia_People) and save

```python
dbpedia_People = dbpedia_ontology + dbpedia_People
#DBpedia_dataset.serialize(destination=input_path + 'DBpedia_dataset.ttl', format='turtle')
dbpedia_People.serialize(destination=input_path + 'dbpedia_People_dataset.nt', format='nt')
```

## Merging graphs (dbpedia_ontology + drugbank) and save

```python
drugbank = dbpedia_ontology + drugbank
#DBpedia_dataset.serialize(destination=input_path + 'DBpedia_dataset.ttl', format='turtle')
drugbank.serialize(destination=input_path + 'drugbank_dataset.nt', format='nt')
```

## Create local endpoint (dbpedia_People, drugbank and dbpedia_ontology )

```python
print(len(dbpedia_People))
print(len(drugbank))
dbp_p=LocalEndpoint(dbpedia_People)
drug=LocalEndpoint(drugbank)
dbp_onto=LocalEndpoint(dbpedia_ontology)
```

## dbpedia_People statistics

```python
subject=dbp_p.select("""
   SELECT ?s (COUNT(*) AS ?count) {
      ?s ?p ?o .
      
   } GROUP BY ?s ORDER BY DESC(?count)
""")

properties=dbp_p.select("""
   SELECT ?p (COUNT(*) AS ?count) {
      ?s ?p ?o .
      
   } GROUP BY ?p ORDER BY DESC(?count)
""")

obj = dbp_p.select("""
   SELECT ?o (COUNT(*) AS ?count) {
      ?s ?p ?o .
      
   } GROUP BY ?o ORDER BY DESC(?count)
""")
```

```python
subject["count"].plot.pie(figsize=(6,6)).set_ylabel('')
```

```python
properties["count"].plot.pie(figsize=(6,6)).set_ylabel('')
```

```python
obj["count"].plot.pie(figsize=(6,6)).set_ylabel('')
```

```python
subject.to_csv(input_path+'DBpedia_people/Statistics/' + 'subject_statistics.csv', sep='\t')
properties.to_csv(input_path+'DBpedia_people/Statistics/' + 'predicate_statistics.csv', sep='\t')
obj.to_csv(input_path+'DBpedia_people/Statistics/' + 'object_statistics.csv', sep='\t')
```
