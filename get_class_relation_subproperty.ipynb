```python
input_path = "/media/rivas/Data1/Data-mining/Kcap/datasets/Yago10/"

schema = input_path + "yagoSchema.ttl"
dataset = input_path + "yago10.nt"
typ = input_path + "Yago10_Type.nt"
tax = input_path + "yagoTaxonomy.ttl"


y_schema = Graph()
y_schema.parse(schema, format="turtle")
y_dataset = Graph()
y_dataset.parse(dataset, format="nt")
y_typ = Graph()
y_typ.parse(typ, format="nt")
y_tax = Graph()
y_tax.parse(tax, format="turtle")
print(len(y_schema))
print(len(y_dataset))
print(len(y_typ))
print(len(y_tax))
```

```python
list_subClass =[]
subProp = ['http://yago-knowledge.org/resource/happenedIn', 'http://yago-knowledge.org/resource/isLocatedIn']
list_subProp = pd.DataFrame(subProp, columns=['s'])
for p in range(list_subProp.shape[0]):
    pred = '<'+ list_subProp['s'][p] +'>'
    query = """SELECT ?sc (COUNT(*) AS ?cnt) {
      ?s """+pred+""" ?o .
      ?s <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> ?o2 .
      ?o2 <http://www.w3.org/2000/01/rdf-schema#subClassOf> ?sc .
   } GROUP BY ?sc ORDER BY DESC(?cnt)"""
    entities = g_yago10.select(query)
    list_subClass.append(entities)
```
