```python
from rdflib import Graph
```

```python
g = Graph()
g.parse("C://DIC/Temp/YAGO/small-example/semantics/steven_spielberg-with-types.nt", format="nt")
    
print(len(g)) # prints 2

#query to get the framework/standard from the sto.nt file
#to get standards we have to replace sto:StandardizationFramework by sto:Standard in the query
#we can get standards of the frameworks as well by just changing the query

qres = g.query(
    """PREFIX owl: <http://www.w3.org/2002/07/owl#>
       PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
       PREFIX ygo: <http://yago-knowledge.org/resource/> 
    
    select ?s where {
            ?s rdf:type ygo:wordnet_person_100007846 .
            } limit 1000""")
    
for row in qres:
    print(row)
```

```python
g = Graph()
g.parse("C://DIC/Temp/YAGO/small-example/semantics/steven_spielberg-with-types.nt", format="nt")
g.parse("C://DIC/Temp/YAGO/small-example/semantics/yagoTaxonomy.ttl", format="ttl")
    
print(len(g)) # prints 2

#query to get the framework/standard from the sto.nt file
#to get standards we have to replace sto:StandardizationFramework by sto:Standard in the query
#we can get standards of the frameworks as well by just changing the query

qres = g.query(
    """PREFIX owl: <http://www.w3.org/2002/07/owl#>
       PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
       PREFIX ygo: <http://yago-knowledge.org/resource/> 
       PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    
    select ?sc where {
            ?s rdf:type ?t .
            ?t rdfs:subClassOf* ?sc .
            } limit 1000""")
    
for row in qres:
    print(row)
```
